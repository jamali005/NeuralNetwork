```python
import time
import math
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# we create a class rectangle that will represent our concepts
class Rectangle:
    def __init__(self, min_x, min_y, max_x, max_y):
        self.min_x = min_x
        self.max_x = max_x
        self.min_y = min_y
        self.max_y = max_y    
    # clissify points
    # function to return labels of some points accordingly to a concept (rectangle)
    def inside_rectangle(self, X):
        labels = np.array((X[:,0] >= self.min_x) & (X[:,0] <= self.max_x) & (X[:,1] >= self.min_y) & (X[:,1] <= self.max_y))
        return labels
    
    
    
def estimate_probability_of_unknown_target_concept(X, R_target):
    # classify our data
    labels = R_target.inside_rectangle(X)

    # Estimate probability of unknown target concept 
    p = np.mean(labels)

    return p

def generate_target_rectangle(X, eps):

    while True:
        min_x = np.min(X[:,0])
        min_y = np.min(X[:,1])
        mid_x = np.median(X[:,0])
        mid_y = np.median(X[:,1])
        max_x = np.max(X[:,0])
        max_y = np.max(X[:,1])
        
        # Generate a target rectangle on selecting randomly a value within the intervals min-max
        rect_min_x = random.randrange(int(min_x), int(mid_x))
        rect_min_y = random.randrange(int(min_y), int(mid_y))
        rect_max_x = random.randrange(int(mid_x), int(max_x))
        rect_max_y = random.randrange(int(mid_y), int(max_y))

        # Generate some target concept
        R_target = Rectangle(min_x=rect_min_x,
                             min_y=rect_min_y,
                             max_x=rect_max_x,
                             max_y=rect_max_y)

        # Estimate probability of unknown target concept 
        p = estimate_probability_of_unknown_target_concept(X, R_target)

        if p >= 3*eps:
            print('Successfully generate concept.')
            return R_target
        else:
            print('Re-generate concept...')


def generate_hypothesis_rectangle(X, R_target):

    # Classify points
    labels = R_target.inside_rectangle(X)

    # Select the tightest corner points in unknown target rectangle
    tight_rect_min_x = np.min(X[labels == 1, 0])
    tight_rect_min_y = np.min(X[labels == 1, 1])
    tight_rect_max_x = np.max(X[labels == 1, 0])
    tight_rect_max_y = np.max(X[labels == 1, 1])

    # Generate hypothesis
    R_hypothesis = Rectangle(min_x=tight_rect_min_x,
                             min_y=tight_rect_min_y,
                             max_x=tight_rect_max_x,
                             max_y=tight_rect_max_y)

    return R_hypothesis

def estimate_generalization_error(X, R_target, R_hypothesis):
    
    # generate true labels for out test data
    true_label = R_target.inside_rectangle(X)

    # make predictions
    predicted_label = R_hypothesis.inside_rectangle(X)

    label_error = (true_label != predicted_label)
    generalization_error = np.mean(label_error)

    return generalization_error
```
